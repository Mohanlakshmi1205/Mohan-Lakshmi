```python
X = dtset.drop(columns="Outcome",axis=1)
Y = dtset["Outcome"]
```

```python
print(X)
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
ss = scaler.transform(X)
print(ss)
```

```python
X = ss
Y = dtset["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
# Train and test the data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y,random_state=30)
```

In train_test_split, random_state controls the "shuffling of the data before splitting it into training and testing sets, ensuring reproducibility by producing the same split each time the code is run with the same integer value".

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# training the model
```

#Training the SVM classifier

```python
Svm = svm.SVC(kernel="linear")
```

```python
Svm.fit(X_train,Y_train)
```

```python
# Evaluate the model
```

```python
# Accuracy score on training data
```

```python
y_pred = Svm.predict(X_train)   # model evaluation on training data
accuracy_score(y_pred,Y_train)
```

an accuracy score (or simply accuracy) is a metric that measures the fraction of correct predictions made by a model, calculated by dividing the number of correct predictions by the total number of predictions.

```python
y_pred1 = Svm.predict(X_test)   # model evaluation on testing data
accuracy_score(y_pred1, Y_test)
```

```python
input = [3,158,76,36,245,31.6,0.851,28]
input_to_numpy_array = np.asarray(input)
input_reshaped = input_to_numpy_array.reshape(1,-1)
std_data = scaler.transform(input_reshaped)
print(std_data)
prediction = Svm.predict(std_data)
print(prediction)
```
